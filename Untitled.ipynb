```python
import os
#word 2 vec
import gensim.models.word2vec as w2v
import gensim
#dimensionality reduction
import sklearn.manifold
#math
import numpy as np
#plotting
import matplotlib.pyplot as plt
#parse dataset
import pandas as pd
#visualization
import seaborn as sns
import sklearn.manifold
from sklearn.decomposition import PCA

#load model
thrones2vec = gensim.models.KeyedVectors.load_word2vec_format(os.path.join("trained", "thrones2vec.w2v"))

#squash dimensionality to 2
tsne = sklearn.manifold.TSNE(n_components=2, random_state=0)

#put it all into a giant matrix
all_word_vectors_matrix = thrones2vec.vectors
pca = PCA(n_components=2)

#train t sne
all_word_vectors_matrix_2d = pca.fit_transform(all_word_vectors_matrix)

#plot point in 2d space
points = pd.DataFrame(
    [
        (word, coords[0], coords[1])
        for word, coords in [
            (word, all_word_vectors_matrix_2d[thrones2vec.wv.vocab[word].index])
            for word in thrones2vec.vocab
        ]
    ],
    columns=["word", "x", "y"]
)
#plot
sns.set_context("poster")
points.plot.scatter("x", "y", s=10, figsize=(20, 12))

def plot_region(x_bounds, y_bounds):
    slice = points.loc[
        (points['x'] >= x_bounds[0]) &
        (points['x'] <= x_bounds[1]) &
        (points['y'] >= y_bounds[0]) &
        (points['y'] <= y_bounds[1])
    ]
    
    ax = slice.plot.scatter("x", "y", s=35, figsize=(10, 8))
    for i, point in slice.iterrows():
        ax.text(point.x, point.y, point.word, fontsize=11)

def nearest_similarity_cosmul(start1, end1, end2):
    similarities = thrones2vec.most_similar_cosmul(
        positive=[end2, start1],
        negative=[end1]
    )
    start2 = similarities[0][0]
    print("{start1} is related to {end1}, as {start2} is related to {end2}".format(**locals()))
    return start2
# Printing the words most closely related to 'Stark'
#print("\n".join(str(word) for word in thrones2vec.most_similar("Stark")))

def plot_region_around_word(word):
    word_point = points[points['word'] == word]
    point = word_point.values
    #Finding the coordinates of the given word and seeing the plot around that value
    plot_region([point[0][1]- 1, point[0][1]+1], [point[0][2]- 1, point[0][2]+1])

plot_region_around_word('Gregor')
```
